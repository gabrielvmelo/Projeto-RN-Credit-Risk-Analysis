# A way to balance dataset
As we know, our dataset is unbalanced, then for our neural network to work properly we need to do something to make our minority class, in this case "class2" bigger. At first we gonna try a variation of SMOTE algorithm, selecting two elements of the minority class and using a simple baricentric equation as (1 - t).P1 + t.P2 = P3 to find a new element that will belong to "class2". That approach is way more faster than the original SMOTE and because of this could be less effective.
Just to compare let's see the class2 atrybute's descriptions.

```python
class2.describe()
```

```python
goal = len(class1)
start = timer()

while len(class2) != goal:
    tam = len(class2)
    rd1 = random.randrange(tam)
    rd2 = random.randrange(tam)

    if rd1 != rd2:
        ex1 = class2.iloc[rd1]
        ex2 = class2.iloc[rd2]
        t = random.randrange(100000)/100000
        ex3 = ((1 - t) * ex1) + (t * ex2)
        ex3.name = tam
        class2 = class2.append(ex3)
        
total = timer() - start
print("{} segundos".format(total))
```
